## Arithmetic Operations on Tensors

### 1. Arithmetic Operator

dimana berisikan sebuah perhitungan arimatika sebuah tensorflow yang ada sekitar 6 opersi aritmetika. yaitu tf.add(), tf.subtraction(), tf.multiply() perkalian tf.divide() pembagian, tf.math.log()logaritma, tf.pow() pangkat

```python
#aritmetika tf.add()
d = tf.constant([[3,5,6,3,1],[4,8,2,6,4],[10,5,2,3,1],
                 [7,6,2,9,2],[5,6,3,4,5]])
e = tf.constant([[1,6,8,9,2],[2,9,5,6,7],[1,2,8,3,6],
                 [2,1,8,9,5],[6,9,10,11,3]])

print("Hasil dari Penjumlahan adalah:\n",tf.add(d, e))
```

```python
#aritmatika tf.multiply()
d = tf.constant([[3,5,6,3,1],[4,8,2,6,4],[10,5,2,3,1],
                 [7,6,2,9,2],[5,6,3,4,5]])
e = tf.constant([[1,6,8,9,2],[2,9,5,6,7],[1,2,8,3,6],
                 [2,1,8,9,5],[6,9,10,11,3]])

print("Hasil dari Perkalian adalah:\n",tf.multiply(d, e))
```

```python
#aritmatika tf.divide()
d = tf.constant([[3,5,6,3,1],[4,8,2,6,4],[10,5,2,3,1],
                 [7,6,2,9,2],[5,6,3,4,5]])
e = tf.constant([[1,6,8,9,2],[2,9,5,6,7],[1,2,8,3,6],
                 [2,1,8,9,5],[6,9,10,11,3]])

print("Hasil dari Pembagian adalah:\n",tf.divide(d, e))
```

```python
#aritmatika tf.pow()
d = tf.constant([[3,5,6,3,1],[4,8,2,6,4],[10,5,2,3,1],
                 [7,6,2,9,2],[5,6,3,4,5]])
e = tf.constant([[1,6,8,9,2],[2,9,5,6,7],[1,2,8,3,6],
                 [2,1,8,9,5],[6,9,10,11,3]])

print("Hasil dari Pangkat adalah:\n",tf.pow(d, e))
```

## Tensor Concatenation and SPlitting

### 1. Tensor Concatenation

#### A. tf.contact()

```python
concat_sample_1 = tf.random.normal([5,100,100,4])
concat_sample_2 = tf.random.normal([50,100,100,4])

print("sizes of the original data:",concat_sample_1.shape,concat_sample_2.shape)
```

```python
concated_sample_1 = tf.concat([concat_sample_1,concat_sample_2],axis=0)
print("size of the concatenated data:",concated_sample_1.shape)
```

#### B. tf.stack()

```python
stack_sample_1 = tf.random.normal([1000,1000,50])
stack_sample_2 = tf.random.normal([1000,1000,50])

print("sizes of the original data: ",stack_sample_1.shape, stack_sample_2.shape)
```

```python
#Dimensions increase after the concatenation. If axis is set to 0, a dimension is added before the first dimension.
stacked_sample_1 = tf.stack([stack_sample_1, stack_sample_2],axis=0)
print("size of the concatenated data:",stacked_sample_1.shape)
```

## Tensor Sorting Methods

### 1. tf.sort() & tf.argsort()

```python
sort_sample_1 = tf.random.shuffle(tf.range(10))
print("input tensor:",sort_sample_1.numpy())
```

```python
sorted_sample_1 = tf.sort(sort_sample_1, direction="DESCENDING")
print("tensor sorted in descending order:",sorted_sample_1.numpy())
```

```python
sorted_sample_1 = tf.sort(sort_sample_1, direction="ASCENDING")
print("tensor sorted in ascending order:",sorted_sample_1.numpy())
```

```python
sorted_sample_2 = tf.argsort(sort_sample_1,direction="ASCENDING")
print("indexes of elements in ascending order:",sorted_sample_2.numpy())
```

```python
sorted_sample_2 = tf.argsort(sort_sample_1,direction="DESCENDING")
print("indexes of elements in descending order:",sorted_sample_2.numpy())
```

```python
sorted_sample_2 = tf.argsort(sort_sample_1,direction="ASCENDING")
print("indexes of elements in ascending order:",sorted_sample_2.numpy())
```

```python
import timeit

cnn_cell = tf.keras.layers.Conv2D(filters=200, kernel_size=2, strides=(1,1))
```

```python
@tf.function
def cnn_fn(image):
    return cnn_cell(image)
```

```python
image = tf.zeros([100,100,100,10])
```

```python
cnn_cell(image)
cnn_fn(image) 
```

```python
print(timeit.timeit(lambda: cnn_cell(image), number=10))
print(timeit.timeit(lambda: cnn_fn(image), number=10))
```
