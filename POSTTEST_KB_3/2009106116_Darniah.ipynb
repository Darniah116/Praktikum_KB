## Grafik Boxplot

```python
plt.boxplot(df[['latitude','menus.amountMin','priceRangeMin']], labels=['Garis Lintang','menu.jumlahMin','harga.minimal'])
plt.title('GRAFIK BOXPLOT')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

## Grafik Heatmap

```python
sns.heatmap(data=df.corr(), cmap ='Greens')
plt.title('GRAFIK HEATMAP')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

## Grafik Batang

```python
df['priceRangeMin'].value_counts().plot.bar()
plt.title('Grafik Batang')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Diagram Batang')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.hist(df['priceRangeMin'], label ='Harga.minimal')
plt.hist(df['menus.amountMax'], label ='menu.jumlahMin')
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

## Grafik Pie

```python
plt.figure(figsize=(10,5))
plt.title('DIAGRAM PIE')
plt.pie(df['priceRangeMin'].value_counts(), labels=['25','50','0','31','7','10','164','347'], startangle=90, autopct='%.1f%%', explode=(0.10,0.10,0.10,0.10,0.10,0.10,0.10,0.10))
plt.show()
```

## Grafik Plot & Garis

```python
plt.figure(figsize=(15,15))
plt.title('Diagram Scatter')
plt.scatter(df['menus.dateSeen'],df['priceRangeMax'], label='HargaMaksimal')
plt.xticks(rotation=90)
plt.xlabel('Tahun')
plt.ylabel('HargaMaksimal')
plt.legend()
plt.show()
```
